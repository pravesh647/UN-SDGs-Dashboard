```python
# Linear Algebra and 
import numpy as np

# Database
import pandas as pd

# visualization
import matplotlib.pyplot as plt

# getting data
import requests
from pandas_datareader import wb

# Web application
# pip install streamlit
import streamlit as st
```

```python
undernourishment="SI.POV.NAHC"
```

```python
year = "2014"
df1=wb.download(indicator=undernourishment, country="all", start=year,end=year).reset_index()
df1 = df1.drop("year", axis =1).rename(columns = {undernourishment:year})
```

```python
df1.dropna()
```

```python


year = "2015"
df2=wb.download(indicator=undernourishment, country="all", start=year,end=year).reset_index()
df2 = df2.drop("year", axis =1).rename(columns = {undernourishment:year})

year = "2016"
df3=wb.download(indicator=undernourishment, country="all", start=year,end=year).reset_index()
df3 = df3.drop("year", axis =1).rename(columns = {undernourishment:year})

year = "2017"
df4=wb.download(indicator=undernourishment, country="all", start=year,end=year).reset_index()
df4 = df4.drop("year", axis =1).rename(columns = {undernourishment:year})

year = "2018"
df5=wb.download(indicator=undernourishment, country="all", start=year,end=year).reset_index()
df5 = df5.drop("year", axis =1).rename(columns = {undernourishment:year})

year = "2019"
df6=wb.download(indicator=undernourishment, country="all", start=year,end=year).reset_index()
df6 = df6.drop("year", axis =1).rename(columns = {undernourishment:year})
```

```python
hunger = df1.merge(df2).merge(df3).merge(df4).merge(df5).merge(df6).set_index("country").dropna()
```

```python
hunger.sort_values("2019", ascending = False)
```
